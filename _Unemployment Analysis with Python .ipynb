```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = {
    "State": ["StateA", "StateA", "StateA", "StateB", "StateB", "StateB"],
    "Date": ["2019-01-01", "2020-06-01", "2021-01-01", "2019-01-01", "2020-06-01", "2021-06-01"],
    "Frequency": ["Monthly"]*6,
    "UnemploymentRate": [5.0, 12.3, 6.0, 4.5, 10.0, 8.0],
    "Employed": [95.0, 87.7, 94.0, 95.5, 90.0, 92.0],
    "LabourParticipation": [60, 58, 61, 62, 55, 59],
    "Year": [2019, 2020, 2021, 2019, 2020, 2021],
    "Month": [1, 6, 1, 1, 6, 6]
}

df = pd.DataFrame(data)

df["Date"] = pd.to_datetime(df["Date"])
```

```python
 df["Date"] = pd.to_datetime(df["Date"])
```

```python
print("First few rows:")
print(df.head())
```

```python
print("\nSummary statistics:")
print(df.describe())

print("\nMissing values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x="Date", y="UnemploymentRate", hue="State",marker="o", data=df)
plt.title("Unemployment Rate Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Year", y="UnemploymentRate", hue="State", data=df)
plt.title("Unemployment Rate Before, During & After COVID-19")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Month", y="UnemploymentRate", data=df)
plt.title("Seasonal Trend in Unemployment Rates (by Month)")
plt.show()
```

```python
 plt.figure(figsize=(5,4))
sns.heatmap(df[["UnemploymentRate", "Employed", "LabourParticipation"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Employment Factors")
plt.show()
```

```python
print("\n Insights:")
print("1. Unemployment spiked in 2020 (COVID-19 period).")
print("2. Both states show recovery in 2021, but not to pre-COVID levels.")
print("3. StateA was more affected than StateB (12.3% vs 10% peak).")
print("4. Labour participation dropped in 2020, then improved in 2021.")
```
